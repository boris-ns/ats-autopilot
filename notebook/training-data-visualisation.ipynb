```python
from keras.models import load_model
import matplotlib.pyplot as plt
import numpy
import pandas as pd

%matplotlib inline 
```

```python
model = load_model("../models/autopilot_new.h5")
training_log = pd.read_csv('../models/training.log', sep=',', engine='python')
```

### History for accuracy

```python
plt.plot(training_log['accuracy'])
plt.plot(training_log['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

### History for loss

```python
plt.plot(training_log['loss'])
plt.plot(training_log['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
